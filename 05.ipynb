```python
%matplotlib inline 
import matplotlib.pyplot as plt
from skimage import io, data, feature, color
import numpy as np

I1 = feature.canny(color.rgb2gray(io.imread('images/haus1.jpg')), 2)
I2 = feature.canny(color.rgb2gray(io.imread('images/haus2.jpg')), 2)
I3 = feature.canny(color.rgb2gray(io.imread('images/haus3.jpg')), 2)
I4 = feature.canny(color.rgb2gray(io.imread('images/haus4.jpg')), 2)

fig = plt.figure(figsize=(15, 10))
ax1 = plt.subplot(2, 2, 1)
ax2 = plt.subplot(2, 2, 2)
ax3 = plt.subplot(2, 2, 3)
ax4 = plt.subplot(2, 2, 4)

ax1.imshow(I1)
ax2.imshow(I2)
ax3.imshow(I3)
ax4.imshow(I4)
```
